```python
df_train=train.rename(columns={'Date':'LagWeek','Weekly_Sales':'Lag_Sales'})
df_train['LagWeek']=pd.to_datetime(df_train['LagWeek'])
df_train.dtypes

df_main=df_main.merge(df_train,how='left',on=['Store','Dept','LagWeek'])
```

```python
#Replace missing values in Lag_Sales by median of the Weekly Sales
df_median=train
df_median['Date']=pd.to_datetime(df_median['Date'])
df_median['Month']=[date.month for date in df_median['Date']]
#df_median.head()
df_median_new=df_median.groupby(['Store','Dept','Month'])['Weekly_Sales'].median().reset_index()
df_median_new.rename(columns={'Weekly_Sales':'Median_Sales'},inplace=True)
df_median_new.head()
```

```python
df_main['Month']=[date.month for date in df_main['Date']]
df_main=df_main.merge(df_median_new,how='left',on=['Store','Dept','Month'])
```

```python
df_main['Lag_available']=np.where(df_main['Lag_Sales'].isna(),0,1)
df_main['Lag_Sales']=np.where(df_main['Lag_Sales'].isna(),df_main['Median_Sales'],df_main['Lag_Sales'])

df_main.drop(['Month','Median_Sales'],axis=1,inplace=True)
df_main.head()
```

```python
df_main_features=df_main.drop(['Weekly_Sales','Lag_Sales'],axis=1)
df_sales=df_main['Weekly_Sales']
```

```python
import datetime as dt
df_main['Date']=df_main['Date'].map(dt.datetime.toordinal)
df_main['LagWeek']=df_main['LagWeek'].map(dt.datetime.toordinal)
```

Because the data has numerical fields, it is important to scale them before feeding it to a model else the model could give biased results. Robus Scaler was tried first since there were outliers but after observing the results, I realised that MinMax Scaler did a better job and the models gave better results.

```python
#from sklearn.preprocessing import RobustScaler

#scaler=RobustScaler()
#numerical=['Size','Temperature','Fuel_Price','Unemployment']

#df_scaled_data=pd.DataFrame(data=df_main_features)
#df_scaled_data[numerical]=scaler.fit_transform(df_scaled_data[numerical])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler(feature_range=(0,1),copy=True)
numerical=['Size','Temperature','Fuel_Price','Unemployment']

df_scaled_data=pd.DataFrame(data=df_main_features)
df_scaled_data[numerical]=scaler.fit_transform(df_scaled_data[numerical])
```

```python
df_scaled_data.head()
```

```python
from sklearn.cross_validation import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(df_scaled_data,df_sales,test_size=0.2,random_state=0)

print ("Training set has {} samples.".format(X_train.shape[0]))
print ("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
X_test['Date']=pd.to_datetime(X_test['Date'])
X_train['Date']=pd.to_datetime(X_train['Date'])
X_train['LagWeek']=pd.to_datetime(X_train['LagWeek'])
X_test['LagWeek']=pd.to_datetime(X_test['LagWeek'])
```

```python
# Data Frame to be able to visulaize the Actuals and Predicted Sales trend over a particular period of time and for one store and one department
df_new=pd.concat([X_test,Y_test],axis=1)

df_new.head()
```

```python
X_train['Date']=X_train['Date'].map(dt.datetime.toordinal)
X_train['LagWeek']=X_train['LagWeek'].map(dt.datetime.toordinal)
X_test['Date']=X_test['Date'].map(dt.datetime.toordinal)
X_test['LagWeek']=X_test['LagWeek'].map(dt.datetime.toordinal)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error, mean_squared_error

lr=LinearRegression()
lr.fit(X_train,Y_train)
pred=lr.predict(X_test)


print("RMSE: {}".format(np.sqrt(mean_squared_error(Y_test,pred))))
print("MAE: {}".format(mean_absolute_error(Y_test,pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train,Y_train)
pred_rf=rf.predict(X_test)

print("RMSE: {}".format(np.sqrt(mean_squared_error(Y_test,pred_rf))))
print("MAE: {}".format(mean_absolute_error(Y_test,pred_rf)))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor(n_neighbors=5)
knr.fit(X_train,Y_train)
pred_knr=knr.predict(X_test)

print("RMSE: {}".format(np.sqrt(mean_squared_error(Y_test,pred_knr))))
print("MAE: {}".format(mean_absolute_error(Y_test,pred_knr)))
```
